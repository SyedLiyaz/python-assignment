```python
!pip install pdfplumber

import os
import pdfplumber
import re
from collections import defaultdict
from google.colab import files

def extract_text_from_pdf(pdf_path):
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text() + "\n"

    print("------ Text Extracted ------")
    return text

def analyze_text(text):
    categories = {
        'growth': ['growth', 'expand', 'increase', 'guidance', 'outlook'],
        'acquisitions': ['acquisition', 'acquire', 'merge', 'synergy'],
        'financial_metrics': ['revenue', 'EBITDA', 'margin', 'PAT', 'ROCE', 'EPS'],
        'customers': ['customer', 'client', 'OEM', 'wins'],
        'risks': ['risk', 'delay', 'dependence', 'slowdown']
    }

    patterns = {
        'percentage': r'\b(\d+\.?\d*%)\b',
        'currency': r'Rs\.\s*(\d+[\.,]?\d*)',
        'quarters': r'Q\d\sFY\d{2}-\d{2}'
    }

    results = defaultdict(list)
    lines = text.split('\n')

    for line in lines:
        line_clean = line.strip().lower()
        for category, keywords in categories.items():
            if any(keyword in line_clean for keyword in keywords):
                context = {
                    'text': line.strip(),
                    'metrics': {}
                }
                percentages = re.findall(patterns['percentage'], line, re.IGNORECASE)
                currencies = re.findall(patterns['currency'], line)
                quarters = re.findall(patterns['quarters'], line)
                if percentages:
                    context['metrics']['percentages'] = percentages
                if currencies:
                    context['metrics']['currency'] = currencies
                if quarters:
                    context['metrics']['quarters'] = quarters
                results[category].append(context)
                break
    return results

def summarize_results(analysis):
    summary = {}
    for category, entries in analysis.items():
        summary[category] = []
        for entry in entries:
            summary_entry = {
                'snippet': entry['text'],
                'metrics': entry['metrics']
            }
            summary[category].append(summary_entry)
    return summary

def main(pdf_path):
    text = extract_text_from_pdf(pdf_path)
    analysis = analyze_text(text)
    summary = summarize_results(analysis)
    return summary


print("Please upload your PDF file.")
uploaded = files.upload()

pdf_path = list(uploaded.keys())[0]

result = main(pdf_path)

for category, entries in result.items():
    print(f"\n=== {category.upper()} ===")
    for idx, entry in enumerate(entries, 1):
        print(f"{idx}. {entry['snippet']}")
        if entry['metrics']:
            print(f"   Metrics: {entry['metrics']}")
```
